```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os 
import cv_scratch as myLib
import time 

```

## z-Norm Data

#### L2 Regularization

```python
X_train = pd.read_csv("data/zNorm/X_train.csv")
X_test = pd.read_csv("data/zNorm/X_test.csv")
y_train = pd.read_csv("data/zNorm/y_train.csv")
y_test = pd.read_csv("data/zNorm/y_test.csv")
X_all = pd.concat([X_train, X_test])
y_all = pd.concat([y_train, y_test])
print("Doing 10-fold Cross-Validation")
all_scores, mean_scores, error_scores = myLib.cv_test(X_all, y_all, k=10)
print("Classifier result: ")
myLib.run_model(X_train, X_test, y_train, y_test, regu='l2')
```
